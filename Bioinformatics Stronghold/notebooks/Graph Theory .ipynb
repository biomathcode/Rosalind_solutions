```python
import networkx as nx
import numpy as np
```

```python
# Initialising an empty graph object

P = nx.Graph() # Create an empty object 
```

```python
# adding nodes 
P.add_nodes_from(['A', 'B', 'C', 'D', 'E'])
```

```python
# adding pairwise relationship
P.add_edges_from([('B','C'),('A', 'C'), ('B', 'D'), ('D', 'A'), ('D', 'E'),('B', 'E')])
```

```python
print(P.nodes())
```

```python
print(P.edges())
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
nx.draw(P, with_labels = True)
```

```python
# Generate the graph
n = 50
p = 0.2
G_erdos = nx.erdos_renyi_graph(n,p, seed =100)
# Plot the graph
plt.figure(figsize=(12,8))
nx.draw(G_erdos, node_size=10)
```

```python
degree_freq = np.array(nx.degree_histogram(G_erdos)).astype('float')
plt.figure(figsize=(12, 8))
plt.stem(degree_freq)
plt.ylabel("Frequence")
plt.xlabel("Degree")
plt.show()
```
